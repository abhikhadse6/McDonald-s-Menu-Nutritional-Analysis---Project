```python
# Define a function to detect outliers based on the IQR method
def detect_outliers_iqr(data_column):
    Q1 = data_column.quantile(0.25)
    Q3 = data_column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data_column[(data_column < lower_bound) | (data_column > upper_bound)]

# Detect outliers for Calories
calorie_outliers = detect_outliers_iqr(data['Calories'])

# Output outliers
calorie_outliers
```

```python
# Remove outliers from the dataset
data_no_outliers = data[~data['Calories'].isin(calorie_outliers)]
```

```python
# Cap outliers at the lower and upper bounds
def cap_outliers(data_column):
    Q1 = data_column.quantile(0.25)
    Q3 = data_column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data_column.clip(lower_bound, upper_bound)

# Apply capping to the Calories column
data['Calories'] = cap_outliers(data['Calories'])
```

## Summary of Findings and Insights


### 1. Calorie Distribution
Highest Calorie Items: The menu items with the highest calorie counts were primarily burgers, breakfast meals, and desserts. Notably, items like the "Big Breakfast with Hotcakes" (750 calories) and the "Double Quarter Pounder with Cheese" (720 calories) stood out as some of the most calorie-dense items. Large-sized desserts like the "McFlurry with M&Ms" also contributed significantly to high calorie intake.
Lowest Calorie Items: Items like the "Side Salad" (15 calories) and "Apple Slices" (15 calories) had the lowest calorie counts. These items offer low-calorie alternatives for customers looking for lighter meal or snack options.
Key Insight: High-calorie items are primarily found in the burger and breakfast categories, while low-calorie options come from salads and fruit sides. This insight can help guide customers looking for either indulgent or lighter meal options.

### 2. Macronutrient Analysis
Fat Content: High-calorie menu items, especially burgers and breakfast items, were also high in fat. The "Double Quarter Pounder with Cheese" and "Big Mac" had significant fat content, making them less suitable for individuals seeking low-fat meals.
Protein Content: Protein was found to be highest in burgers and chicken-based items. Grilled chicken sandwiches and McNuggets provided relatively high protein content while still maintaining moderate calorie counts, making them attractive to customers seeking high-protein options.
Carbohydrates: Items such as the "Big Breakfast with Hotcakes" and large fries exhibited high carbohydrate content, as did the dessert items like milkshakes and McFlurries.
Key Insight: Burgers and breakfast items are rich in protein but also high in fat and calories. For protein-rich but lower-calorie options, customers could choose grilled chicken sandwiches. Desserts, on the other hand, contribute significantly to carbohydrate intake.

### 3. Healthiest and Least Healthy Options
Healthier Options: The healthier items, based on calorie, fat, and nutrient balance, were primarily salads and fruit sides. These items have a low caloric load and minimal fat content. Salads such as the "Side Salad" (15 calories) and fruit items like "Apple Slices" (15 calories) are excellent choices for health-conscious consumers.
Less Healthy Options: The less healthy options, characterized by high calorie, fat, and sugar content, include the "Big Breakfast with Hotcakes" and the "Double Quarter Pounder with Cheese." These items are high in saturated fat and total calories, making them less ideal for regular consumption.
Key Insight: Offering clear distinctions between healthy and less healthy options can empower customers to make better choices. Highlighting healthier items on the menu and in promotions could encourage customers to select more nutritious meals.

### 4. Category-wise Nutritional Content
Burgers and Sandwiches: This category tends to be calorie-dense, with many items surpassing 400-600 calories. However, they also offer higher protein content.
Salads and Sides: This category provides lighter, lower-calorie options, making it suitable for those watching calorie intake. Salads, in particular, have very low calorie and fat content while still offering some fiber and nutrients.
Desserts and Beverages: These items often contain high sugar and carbohydrate content, contributing to excess calorie consumption. McFlurries and large milkshakes are the primary offenders in this category.
Key Insight: Different categories offer different nutritional profiles, and categorizing menu items based on health metrics (calories, protein, fat) can help guide customer choices.

### 5. Recommendations for McDonald’s
Introduce More Low-Calorie Options: Expanding the salad, fruit, and vegetable offerings will cater to health-conscious customers.
Promote Protein-Rich, Lower Calorie Meals: Highlighting options like grilled chicken sandwiches that provide protein without excessive calories or fat could appeal to fitness-focused customers.
Portion Control for High-Calorie Items: Offering smaller portions of high-calorie items (e.g., smaller fries, mini-burgers) could give customers indulgent options in controlled quantities.
